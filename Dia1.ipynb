# Compare Python List & NumPy Array

```python
# Create two python list say 'list_x' and 'list_y'
list_x = list(range(100000))
list_y = list(range(100000))
```

```python
# Create two numpy array (say 'arr_x' and 'arr_y') using above two list
arr_x = np.array(list_x)
arr_y = np.array(list_y)
```

```python
# Define a function to calculate sum using pythom lists (list_x, list_y)
def list_sum():
    z = [] # a empty list to store sum
    for i in range(len(list_x)):
        z.append(list_x[i] + list_y[i])
```

```python
# Define a function to calculate sum using numpy array (arr_x, arr_y)
def numpy_arr_sum():
    # In numpy we can directñy perform sum element-wise as
    z = arr_x + arr_y
```

```python
# Import Timer class from timeit library
from timeit import Timer
```

```python
timer_obj1 = Timer("list_sum()", "from __main__ import list_sum")
timer_obj2 = Timer("numpy_arr_sum()", "from __main__ import numpy_arr_sum")

print("Pure python version:", timer_obj1.timeit(1000))
print("Numpy version:", timer_obj2.timeit(1000))
```

Everything in Python is an object.

Suppose you have an object and want to know what type of object it is; you will use the type() function to get this information.

## Caution!!
A NumPy array doesn’t support the append() method directly on the array. The given snippet will clear this:



```python
# NOte: NumPy array doesn't suppert append directly on the array
arr7.append(20)
```
